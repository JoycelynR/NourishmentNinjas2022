```python
        # imports
from tkinter import *
from PIL import ImageTk,Image
import pandas as pd
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,
NavigationToolbar2Tk)


        # declare
root = Tk()
root.title('Nourishment Ninjas')

def open():
    window2 = Toplevel(root)
    window2.title("new window")
    lbl = Label(window2, text="this is new window")
    lbl.pack()
  
def show():
    myLabel = Label(root, text=clicked.get()).pack()
def move():
    window1 = Toplevel(root)
    window1.title("Graph")
    lbl1 = Label(window1, text="Graphical Data")
    lbl1.pack()
def plot():
    data = pd.read_csv('NC_FoodSource_Data.csv', low_memory=False)
    data.groupby('Type').count()[['Address']]

# the main Tkinter window
window2 = Tk()

# setting the title
window2.title('Plotting in Tkinter')

# dimensions of the main window
window2.geometry("500x500")

# button that displays the plot
plot_button = Button(master = window2,
					command = plot,
					height = 2,
					width = 10,
					text = "Food Insecurity by County")

# place the button
# in main window
plot_button.pack()

# run the gui
window2.mainloop() 
    
    
        # open screen
frame = LabelFrame(root, text = "WELCOME TO YOUR FOOD INSECURITY RESOURCE", padx=25, pady=25, fg ="white", bg="grey")
frame.pack(padx=25, pady=25)



        # button 1
btn = Button(frame, text= "Nourishment Ninjas", padx=30, pady=30, fg ="green", bg="black",command=open)
btn.grid(row=1, column=1)
btn.config(font=('Ink Free',20,'bold'))

        # menu

options = [
    "Option 1 - OVERVIEW",
    "Option 2 - HEALTHCARE",
    "Option 3 -",
]

clicked = StringVar()
clicked.set(options[0])

drop = OptionMenu(root, clicked, *options)
drop.pack()

showBtn = Button(root, text= "Show Selected Option", command=move).pack()





root.mainloop()
```
